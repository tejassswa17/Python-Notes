### Repeat with while

The while statement in Python supports repeated execution of a statement or block of statements that is controlled by a conditional expression.
```
x = 5
while x > 0:
  x -= 1           
```

```
while x > 0:
```
First, expression, which is known as the loop condition, is evaluated.
* If the condition is false, the while statement ends.
* If the loop condition is satisfied, the statement or statements that comprise the loop body are executed.

```
while x > 0:
  x -= 1            
```
**When the loop body finishes executing, the loop condition is evaluated again, to see if another iteration should be performed.**
* This process continues until the loop condition is false, at which point the while statement ends.


```
x -= 1
```

* The loop body should contain code that eventually makes the loop condition false, or the loop will never end unless an exception is raised or the loop body executes a break statement.

* The while loop in python is used to run a block of code as long as the condition being checked for is True

 **A loop that is in a function’s body also ends if a return statement executes in the loop body, as the whole function ends in this case.**

Syntax:
```
while condition:
    block_of_code
    i += 1 # i += 1
```

1. Evaluate the condition, yielding True or False.
2. If the condition is false, exit the while statement and continue execution at
the next statement.
3. If the condition is true, execute the body and then go back to step 1.

**This type of flow is called a loop because the third step loops back around to the top. We call each time we execute the body of the loop an iteration.**

### Transfer Statements

Transfer statements in python are used to control the flow of the loop and to alter its course. There are three control flow statements:
- `pass`: The pass keyword is used in the place of a block which is required syntactically but not logically.
- `break`: The break keyword is used to terminate the loop altogether
- `continue`: The continue keyword is used to terminate the current iteration

#### pass

````python
# writing a conditional where we know what to do if the condition is false but
# the reverse is not true

a = 100
if a == 10:
    pass
else:
    print('The value is not 10')
````

### Iterate with for

The for statement in Python supports repeated execution of a statement or block of statements that is controlled by an iterable expression.

    for target/element in iterable:

Note that the in keyword is part of the syntax of the for statement and is functionally unrelated to the in operator used for membership testing.

**Key Differences Between While Loops and For Loops**

| Feature                         | `while` Loop                                | `for` Loop                                     |
|---------------------------------|---------------------------------------------|------------------------------------------------|
| **Loop Duration**              | Continues as long as a condition is true     | Runs a **fixed number** of iterations          |
| **Initialization**             | You manually set up any counter or variable  | Counter is defined within the loop header      |
| **Increment/Update**           | Must be handled inside the loop manually     | Automatically handled by loop structure        |
| **Use Case**                   | Useful when the end condition is **unknown** | Ideal when the number of iterations is **known**|

#### Generate Number Sequences with range()

**The range() function returns a stream of numbers within a specified range. without first having to create and store a large data structure such as a list or tuple. This lets you create huge ranges without using all the memory in your computer and crashing your program.**

A range object is an iterable that yields integers sequentially on demand—one at a time—without ever loading the entire sequence into memory. This makes it memory-efficient. Although it represents a definite (finite) set of integers, the values are generated lazily as you iterate through the range.

Looping over a sequence of integers is a common task, so Python provides built-in functions range to generate and return integer sequences. The simplest, most idiomatic way to loop n times in Python is:


    for i in range(n):
      statement(s)

* range( arguments ) returns an iterable object.
* range( x ) returns a list whose items are consecutive integers from 0 (included) up to x (excluded).
* range( x , y ) returns a list whose items are consecutive integers from x (included) up to y (excluded). The result is the empty list if x is greater than or equal to y.
* range( x , y , step ) returns a list of integers from x (included) up to y (excluded), such that the difference between each two adjacent items in the list is step.

### Comprehensions
- Comprehensions offer an easy and compact way of creating lists,
sets and dictionaries.
- A comprehension works by looping or iterating over items and
assigning them to a container like list, set or dictionary.
- This container cannot be a tuple as tuple being immutable is unable
to receive assignments.
- The expression form, called a comprehension, because it lets you code this common idiom concisely and directly.
- Since comprehension is an expression (rather than a block of statements), you can use it directly wherever you need an expression (e.g., as an actual argument in a function call, in a return statement, or as a subexpression for some other expression).

#### List Comprehension
List comprehension consists of brackets containing an expression **followed by a for clause, and zero or more for or if clauses**. So general form of a list comprehension is

`lst = [expression for var in sequence [optional for and/or if]]`


````python
# convert a list of strings to a list of integers
b = []
a = ['10', '20', '30', '40']
for ele in a:
  b.append(int(ele))
b
````

````python
# convert a list of strings to a list of integers
a = [int(x) for x in ['10', '20', '30', '40']]
a
````

````python
# using for loop
a = [1,2,3,4]
b = []
for i in a: # iteration
    b.append(i**2) # expression
b
````

````python
# using list comprehensions
a = [1,2,3,4]
a
b = [i** 2 for i in a]
b
````

**The for clause in a list comprehension can be followed by one or more if and
for statements that are applied to the values produced by the for clause.**

````python
# generate a list of even numbers in the range 10 to 30
b = []
for ele in range(10, 31): # iteration
  if ele%2 == 0: # filter
    b.append(ele) # expression
b
````

````python
# generate a list of even numbers in the range 10 to 30
a = [n for n in range(10, 31) if n % 2 == 0]
a
````

````python
# use of if in list comprehension
a = [1,2,3,4]
b = [i** 2 for i in a if i > 2]
b
````

````python
#  use of if-else in list comprehension
a = [1,2,3,4]
b = [i** 2 if i > 2 else None for i in a]
b
````

It consists of braces containing an expression followed by a for clause, and zero or more for or if clauses. So general form of a set comprehension is

`s = {expression for var in sequence [optional for and/or if]}`


````python
# generate a set containing square of all numbers between 0 and 10
a = {x**2 for x in range(10)}
a
````

````python
# Above comprehension is similar to
a = set()
for x in range(10):
  a.add(x**2)
a
````

````python
# from a set delete all numbers between 20 and 50
a = {num for num in a if num > 20 and num < 50}
a
````

````python
# Above comprehension is similar to
out = set()
for num in a:
  if num > 20 and num < 50:
    out.add(num)
out
````

#### Dictionary Comprehension
General form of a dictionary comprehension is as follows:

`dict_var = {key:value for (key, value) in dictonary.items( )}`

````python
d = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
# obtain dictionary with each value cubed
d1 = {k : v ** 3 for (k, v) in d.items()}
print(d1)
````

````python
# Above comprehension is similar to
d1 = {}
for k,v in d.items():
  d1[k] = v**3
d1
````

````python
# obtain dictionary with each value cubed if value > 3
d2 = {k : v ** 3 for (k, v) in d.items( ) if v > 3}
print(d2)
````

````python
# Above comprehension is similar to
d2 = {}
for k, v in d.items():
  if v > 3:
    d2[k] = v**3
d2
````

````python
# Identify odd and even entries in the dictionary
d3 = {k : ('Even' if v % 2 == 0 else 'Odd') for (k, v) in d.items( )}
print(d3)
````

````python
# Above comprehension is similar to
d3 = {}
for k, v in d.items():
  if v % 2 == 0:
    d3[k] = 'Even'
  else:
    d3[k] = 'Odd'
d3
````

````python
# Using dictionary comprehensions
t = {k:"pass" if v>=50 else "fails" for k,v in d.items()}
t
````

````python
# Above comprehension is similar to
d = {"milan":17 , "saran":85,"nikil":78}
t = {}
for k, v in d.items():
    grade = "pass" if v >=50 else "fails"
    t[k] = grade
print(t)
````

#### There Are No Tuple Comprehensions - only generator object
* Mutable types (lists, dictionaries, and sets) have comprehensions. Immutable types like strings and tuples need to be created with the other methods listed in their sections.
* You might have thought that changing the square brackets of a list comprehension to parentheses would create a tuple comprehension.

````python
number_thing = (number for number in range(1, 6))
number_thing
````

The thing between the parentheses is something else entirely: a generator comprehension, and it returns a generator object

````python
type(number_thing)
````

````python
for ele in number_thing:
  print(ele)
````

````python
i = 0
while i < 10:
  print(i)
  i += 1

lst = ['a', 'b', 'c', 'd']
j = 0
while j < len(lst):
  print(lst[j])
  j+=1
````
