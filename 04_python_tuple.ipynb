## Tuples

**What are Tuples?**
- Though a list can store *dissimilar data*, it is commonly used for
storing **similar data**.
- Though a tuple can store *similar data* it is commonly used for storing **dissimilar data**.
- The tuple data is enclosed within ( ) as shown below, but to create a tuple parenthesis is optional and seperator comma is mandatory.
- In comparision with lists tuples are computationally efficient during creation, storage, copy and retreive operations.


```python
a = ( ) # empty tuple
b = ('Sanjay', 25, 34555.50) # tuple with dissimilar items
c = (10, 20, 30, 40) # # tuple with similar items
print('Empty tuple', a)
print('Dissimilat elements tuple', b)
print('Similar elements tuple', c)
```

**While creating the tuple e, if we do not use the comma after 10, e is
treated to be of type int.**

* Since parentheses are used to group expressions, (1) is
merely a verbose way to write the integer 1. To denote the singleton tuple
containing this value, we write (1,).

```python
e = (10) # one integer within parenthesis can be considered as integer
f = (10,) # tuple with one item. , after 10 is necessary
print('One element in parenthesis', e, 'its type is', type(e))
print('One element in parenthesis with a comma after it', f, 'and its type is:', type(f))
```

**While initializing a tuple, we may drop ( ).**

```python
g = 'Sanjay', 25, 34555.50 # tuple with dissimilar items
h = 10, 20, 30, 40 # # tuple with similar items
print('Dissimilat elements tuple', g, 'and its type is:', type(g))
print('Similar elements tuple', h, 'and its type is:', type(h))
```

**Items in a tuple can be repeated, i.e. tuple may contain duplicate
items. However, unlike list, tuple elements cannot be repeated using
a *.**

```python
i = (10)*5 # one integer within parenthesis can be considered as integer
j = (10,)*5 # tuple with one item. , after 10 is necessary
print('One element in parenthesis', i, 'its type is', type(i))
print('One element in parenthesis with a comma after it', j, 'and its type is:', type(j))
```

### Tuple Varieties
* It is possible to create a tuple of tuples.

```python
a = (1, 3, 5, 7, 9)
b = (2, 4, 6, 8, 10)
c = (a, b)
print(c[0][0], c[1][2]) # 0th element of 0th tuple, 2nd ele of 1st tuple
```

**A tuple may be embedded in another tuple.**

```python
x = (1, 2, 3, 4)
y = (10, 20, x, 30)
print(y) # outputs (10, 20, (1, 2, 3, 4), 30)
```

**It is possible to create a list of tuples, or a tuple of lists**

```python
lst = [('Priyanka', 24, 3455.50), ('Shailesh', 25, 4555.50)]
lst
```

```python
tpl = (['Priyanka', 24, 3455.50], ['Shailesh', 25, 4555.50])
tpl
```

**If we wish to sort a list of tuples or tuple of lists, it can be done as
follows**

* By default, sorted( ) sorts by first item in list/tuple, i.e. name.
* If we wish to sort by salary, we need to use the itemgetter( )
function of operator module.
* In general, operator.itemgetter(n) constructs a function that takes a
list/tuple as input, and fetches the n-th element out of it.
* The key parameter of sorted( ) requires a key function (to be applied to objects to be sorted) rather than a single key value.


```python
import operator
# each embedded tuple/list contains name, age,salary
lst = [('Shailesh', 24, 3455.50), ('Priyanka', 25, 4555.50)]
print(sorted(lst))
print(sorted(lst, key = operator.itemgetter(2)))
```

```python
tpl = (['Shailesh', 24, 3455.50], ['Priyanka', 25, 4555.50])
print(sorted(tpl))
print(sorted(tpl, key = operator.itemgetter(2)))
```

### Accessing Tuple Elements

Tuple is an ordered collection. So order of insertion of elements in a
tuple is same as the order of access. So like a string and list, tuple
items too can be accessed using indices, starting with 0.

```python
msg = ('Handle', 'Exceptions', 'Like', 'a', 'boss')
print(msg[0], msg[3], msg[-1])
```

**Like strings and lists, tuples too can be sliced to yield smaller tuples.**

```python
emp = ('Sanjay', 23, 23000, 1760, 2040)
print(emp[1:3]) # prints (23, 23000)
print(emp[3:]) # prints (1760, 2040)
print(emp[:3]) # prints ('Sanjay', 23, 23000)
```

### Basic Tuple Operations

**Mutability - Unlike a list, a tuple is immutable, i.e. it cannot be
modified.**
* Since a tuple is immutable operations like append, remove and
insert do not work with a tuple.

```python
msg = ('Fall', 'In', 'Line')
# msg[0] ='FALL' # TypeError: 'tuple' object does not support item assignment
# msg[1:3] = ('Above', 'Mark') # TypeError: 'tuple' object does not support item assignment
print('Tuple:', msg)
```

**Though a tuple itself is immutable, it can contain mutable objects
like lists**

```python
# mutable lists, immutable string—all can belong to tuple
s = ([1, 2, 3, 4], [4, 5], 'Ocelot')
print('Tuple with mutable lists and immutable strings:', s)
```

**If a tuple contains a list, the list can be modified since list is a
mutable object.**

```python
s = ([1, 2, 3, 4], [10, 20], 'Oynx')
print('Initial tuple', s)
print('Identity of initial tuple', id(s))
s[1][1] = 45 # changes first item of first list, i.e. 20
print('Modified tuple', s) # prints ([1, 2, 3, 4], [4, 45], 'Oynx')
print('Identity of modified tuple', id(s))
```

```python
# one more way to change first item of first list
s = ([1, 2, 3, 4], [10, 20], 'Oynx')
print('Initial tuple', s)
print('Identity of initial tuple', id(s))

p = s[1]
p[1] = 100
print('Modified Tuple', s) # prints ([1, 2, 3, 4], [4, 100], 'Oynx')
print('Identity of modified tuple', id(s))
```

### Tuple unpacking
* It is possible to unpack a tuple within a tuple using the *operator.

```python
x = (1, 2, 3, 4)
y = (10, 20, *x, 30)
print(y) # outputs (10, 20, 1, 2, 3, 4, 30)
```

**Pass a tuple to the divmod( ) function and obtain the quotient and the
remainder. divmod( ) returns a tuple consisting of quotient and remainder.**

```python
tpl = (7,3)
divmod(tpl) # divmod((7, 3))
# TypeError: divmod expected 2 arguments, got 1
```

```python
tpl = (7,3)
divmod(*tpl)
```

```python
tpl = (10, 20, 30, 40, 50, 60, 70, 8, 90, 100)
print(tpl)
```

```python
a, b, c, d, e, f, g, h, i, j = tpl
print(a, b, c, d, e, f, g, h, i, j)
```

**Disposable variable _ is usally used when you do not wish to use the
variable further, and is being used only as a place-holder.**

**Unpack the tuple such that first value gets stored in variable x, last
value in y and all valuesin between into disposable variables _**

```python
x, _, _, _, _, _, _, _, _, y = tpl
x, y, _
```

**Unpack the tuple such that first value gets stored in variable x, last
value in y and all values in between into a single disposable variable _**

```python
x, *_, y = tpl
x, y, _
```

**unpacking a nested data structure**

```python
lst = [1, 2, 3, [4, 5, 6]]
```

```python
a, b, c, (d, e, f) = lst
a, b, c, d, e, f
```

```python
a, b, c, (*d) = lst
a, b, c, d
```

```python
a, b, c, (*d,) = lst
a, b, c, d
```

```python
a, *b, (d, *e) = lst
a, b, d, e
```
