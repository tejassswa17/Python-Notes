## Sets
- Sets are like lists, with an exception that they do not contain
duplicate entries and elements are not in any particular order
- A set in Python is an unordered collection (cannot have indexes) of objects used in situations where membership (You use a set when you only want to know that something exists, and nothing else about it) and uniqueness in the set are main things you need to know about that object.
- A set is like a dictionary with its values thrown away, leaving only the keys. As with a dictionary, each key must be unique. . Use a dictionary if you
want to attach some information to the key as a value.


**While storing an element in a set, its hash value is computed using a
hashing technique to determine where it should be stored in the set.**

```python
# sets declaration
my_set = {1,2,3}
print('Set:', my_set)

my_set1 = {1,"Hello",3.5, 'True', None}
print('Set is:',my_set1)
```


**Since hash value of an element will always be same, no matter in
which order we insert the elements in a set, they get stored in the
same order.**

It is possible to create a set of strings and tuples, but not a set of
lists.

**Since strings and tuples are immutable, their hash value remains
same at all times. Hence a set of strings or tuples is permitted.
However, a list may change, so its hash value may change, hence a
set of lists is not permitted.**

```python
s1 = {'Morning', 'Evening'} # works
s2 = {(12, 23), (15, 25), (17, 34)} # works
# s3 = {[12, 23], [15, 25], [17, 34]} # TypeError: unhashable type: 'list'
s1, s2
```

### Accessing Set Elements
- Entire set can be printed by just using the name of the set. Set is an
unordered collection. Hence order of insertion is not same as the
order of access
- Sets cannot be sliced using [ ].

```python
s = {15, 25, 35, 45, 55}
print(s) # prints {35, 45, 15, 55, 25}
```

Being an unordered collection, items in a set cannot be accessed
using indices.

### Set operations

#### union(|)

```python
#Sets union, intersection and difference
A = { 1,2,3,4,5,6}
B = {5,6,7,8,9,10}

print("A union B", A|B)
print("A union B can also be written as", B|A)
print("Another way of representation",A.union(B))
print("Another way of representation",B.union(A))
```

```python
A |= B
print("A union B", A)
```

#### intersection(&)

```python
#Sets union, intersection and difference
A = { 1,2,3,4,5,6}
B = {5,6,7,8,9,10}

print("A intersection B", A&B)
print("A intersection B can also be written as", B&A)
print("Another way of representation",A.intersection(B))
print("Another way of representation",B.intersection(A))
```

```python
A &= B
print("A intersection B", A)
```

#### difference(-)

```python
# This operation is not commutative
A = { 1,2,3,4,5,6}
B = {5,6,7,8,9,10}

print("A difference B", A - B)
print("A difference B" , A.difference(B))
```

```python
A -= B
print("A difference B", A)
```

#### symmetric difference(^)
* Symmetric difference prints values of A and B and eliminates the common elements from both the sets

```python
# This operation is commutative
A = { 1,2,3,4,5,6}
B = {5,6,7,8,9,10}
print('Symmetric difference between A and B', A ^ B)
print('Symmetric difference between A and B', A.symmetric_difference(B))
print('Symmetric difference between A and B', B.symmetric_difference(A))
```

```python
 A ^= B
 print('Symmetric difference between A and B',A)
```

```python
#Symmetric difference can also be written as below notation
print ((A|B) - (A & B))
```

## Frozen sets
* Because sets aren’t immutable and hashable, they can’t belong to other sets. To remedy that situation there is another set type, frozenset, which is just like a set but can’t be changed after creation.

* Because frozensets are immutable and hashable, they can be members of other sets

* There is no special syntax to represent frozenset literals—they must be created by calling the constructor.

```python
 frozenset([3, 2, 1])
```

```python
frozenset(set([2, 1, 3]))
```

```python
frozenset({3, 1, 2})
```

```python
frozenset( (2, 3, 1) )
```

**Is it really frozen?**

```python
s = frozenset({'gate', 'fate', 'late'})
s.add('rate') # error
```

```python
frozenset(range(10))
```
