### What is a function?

* Function is a block of code(a group of statements) that performs a specific and well defined task.
* There are two types of Python functions:
> 1. Built-in functions - Ex. len( ), sorted( ), min( ), max( ), etc.
> 2. User-defined functions

### What are the advantages with functions?
1. They help us divide our program into multiple tasks. For each
task we can define a function. This makes the code modular.
2. Functions provide a reuse mechanism. The same function can be
called any number of times.

- **`Params/Args`** - used to pass values to a function. Params are optional.
- **`Colon (:)`**- marks the end of the function header.
- **`Doc String`** - A short description about the function.This is optional.
- **`Business logic/Statements`** - One ore more valid Python statements to perform the required task.
- **`return`** - This is optional. But this statement will help you to return a value from the function.
- **`print`** - To dispaly the value from the function. This is optional.
Either print or return need to be included at the end of the function.
Make sure proper indentation is given inside the function body.

###  The Scope of Variables

**Namespace and scope differ in that a namespace maps names to objects, while a scope is the region of code where you can access a name.**

**NOTE**: In this notebook `namespace` and `scope` are used intechangeably.

**Why Scope and namespaces?**

Understanding how namespaces work will improve your ability to manage and organize code efficiently in Python programs, helping to prevent name conflicts and other issues.

In Python, the global, local, and nonlocal namespaces are represented using dictionaries. However, the built-in namespace is different—it’s implemented as the builtins module, not as a dictionary.

* Scope of an identifier indicates where it is available for use.
* Scope can be Local (L), Enclosing or Nonlocal (E), Global (G), Built-in (B). Scope becomes more and more liberal from Local to Built-in. This can be best understood though the programs given below.


**The built-in namespace** holds the identifiers for all of Python’s built-in objects, such as standard functions (like print() and len()), exceptions, and commonly used types (like int and list). This namespace is automatically loaded and remains accessible for the entire duration that the Python interpreter is active. As a result, you can use any built-in function or object at any point in your Python code, without having to explicitly import or define them—Python ensures they’re available globally throughout your program’s execution.

* `len`, `str`, `print` can be used anywhere in the program without
importing any module. So they have a built-in scope.

`````python
print(len("hello"))  # The 'len' function belongs to the built-in namespace.
`````

* This is also called the module-level namespace because **each Python file is a module**. Each module has its own global namespace.

Python creates a main global namespace when the main program’s body starts.

* When you run a Python script directly (e.g., 09_python_functions.ipynb), Python treats that script as the "main module" and creates a global namespace for it.

**This local namespace exists only during the execution of the function. As soon as the function finishes (returns, raises an exception, or exits otherwise), the local namespace is destroyed.**

In Python, you can define functions inside other functions. When you do this, the outer function creates its own namespace that exists as long as the function is executing. If you define an inner function inside an outer function, the inner function has access to the variables of the outer (enclosing) function—that’s what’s called the **enclosing or nonlocal namespace.**

**The namespace of the outer function, accessible to any inner functions nested within it.**

`````python
x = "I am global"

def outer_func():
    nonlocal_variable = "I am nonlocal (enclosing)"

    def inner_func():
        local_variable = "I am local to inner_func"
        print(x)  # accesses global
        print(nonlocal_variable)  # accesses enclosing (nonlocal)
        print(local_variable)  # accesses local

    inner_func()
    # print(local_variable)  # <-- Would give an error; not defined in this scope

outer_func()
# print(nonlocal_variable)  # <-- Would give an error; not defined in this scope
`````

`````python
x = "I am global"

def outer_func():
    nonlocal_variable = "I am nonlocal (enclosing)"

    def inner_func():
        local_variable = "I am local to inner_func"
        print(x)  # accesses global
        print(nonlocal_variable)  # accesses enclosing (nonlocal)
        print(local_variable)  # accesses local

    inner_func()
    print(local_variable)  # <-- Would give an error; not defined in this scope

outer_func()
# print(nonlocal_variable)  # <-- Would give an error; not defined in this scope
`````

`````python
x = "I am global"

def outer_func():
    nonlocal_variable = "I am nonlocal (enclosing)"

    def inner_func():
        local_variable = "I am local to inner_func"
        print(x)  # accesses global
        print(nonlocal_variable)  # accesses enclosing (nonlocal)
        print(local_variable)  # accesses local

    inner_func()
    # print(local_variable)  # <-- Would give an error; not defined in this scope

outer_func()
print(nonlocal_variable)  # <-- Would give an error; not defined in this scope
`````

* `x` lives in the global namespace.

* `nonlocal_variable` is local to outer_func, but nonlocal to inner_func (accessible inside inner_func via closure).

* `local_variable` exists only within inner_func.

* When outer_func() finishes, its namespace is deleted and any variables become invalid, unless they're still referenced by closures.

**What happens to the namespace of a function when it terminates?**

When a function terminates, all references to the objects in its namespace cease to be valid. This means that while Python might not immediately reclaim the memory allocated for those namespaces, the objects within them are no longer accessible.

**How Python Searches for Names???**

**The LEGB rule** describes the precise order in which the Python interpreter looks up variable and function names (identifiers) when evaluating them inside code blocks. LEGB stands for Local, Enclosing, Global, and Built-in—these are the four scopes or namespaces Python searches, in order, when trying to resolve a name.

1. Local (L):

> * The innermost scope.

> * Names created inside the current function, including its parameters and variables assigned within.

> * **Searched first** when a name is referenced inside a function.

2. Enclosing (E):

> * The scope(s) of any enclosing functions, in cases of nested functions.

> * Applies when a function is defined inside another function (“closure”).

> * Python will look in these outer, but non-global, function scopes if not found locally.

3. Global (G):

> * The top-level of the module/file (or interactive session) where execution is happening.

> * Names assigned at the module level (variables, functions, classes).

> * **Searched third**, after exhausting local and enclosing scopes.

4. Built-in (B):

> * Python’s built-in module, builtins, containing standard identifiers like len, print, range, etc.

> * **Searched last.** If the name isn't found here, Python raises a NameError.

### Recursion -  a function calls itself

Recursion means calling a function in terms of iteself. All recursive algorithms have these characteristics:

1. A recursive algorithm must call itself, recursively.
2. A recursive algorithm must have a base case. A base case occurs when the problem after repeated calls to the recursive function, becomes so small that it can be solved directly.
3. A recursive algorithm after each call to the recursive function, must move towards the base case. So, in a well-designed recursive algorithm, after each iteration, the problem must become smaller.

`````python
# Sum of digits of a number
def digit_sum(num):
  r = num%10
  num //= 10
  if r == 0:
    return 0
  else:
    return r + digit_sum(num)
`````

````
quotient, remainder = divmod(dividend, divisor)
````

**Head recursion - In this type of recursion the recursive call is made before other processing in the function.**

`````python
def headprint(n) :
  if n == 0 :
    return 'end'
  else :
    headprint(n - 1)
    print(n)

headprint(10)
`````

**In this type of recursion processing is done before the recursive call. The tail recursion is similar to a loop—the function executes all the statements before making the recursive call.**

`````python
def tailprint(n):
  if n == 0:
    return
  else:
    print(n)
    tailprint(n - 1)

tailprint(10)
`````

#### Filter function
- A filter operation applies a function to all the elements of a sequence. A sequence of those elements for which the function returns True is returned.
- filter() function applies the predicate (condition) on sequence. If the predicate returns True, the value passes the test and is added to a filter object. Otherwise, the value is dropped from consideration.
````
filter(function_to_apply, list_of_inputs)
````

- filter( ) returns a filter object which can be converted to a list using
list( ) function.
