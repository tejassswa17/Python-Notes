### Exceptions
**Exceptions are events that can modify the flow of control through a program.**

*  Even if the program is grammatically correct, things may go wrong
during execution causing exceptions.
* If things go wrong during execution (runtime):
> * **Means** - Something unforeseen has happened
> * **Reported by** - Python Runtime
> * **Action to be taken** - Tackle it on the fly

*In Python, exceptions are triggered automatically on errors, and they can be triggered and intercepted by your code. they are processed by
four statements*
1. **raise** - Trigger an exception manually in your code.
2. **assert** - Conditionally trigger an exception in your code.
3. **try/except** - Catch and recover from exceptions raised by Python, or by you.
4. **try/finally** - Perform cleanup actions, whether exceptions occur or not.

#### Assertion

##### How to use try - except?
* **try block** - Enclose in it the code that you anticipate will cause an
exception.
* **except block** - Catch the raised exception in it. It must immediately
follow the try block.

`````python
a = int(input('Enter an integer: '))
b = int(input('Enter an integer: '))
try:
  div(a, b)
except ZeroDivisionError:
  print('Denominator is 0')
print('Performed operation successfully')
`````

**If the type of exception raised matches the exception named after except keyword, that except block is executed.**

**If no exception occurs while executing the try block, control goes to
first line beyond the except block.**

`````python
try:
  a = int(input('Enter an integer: '))
  b = int(input('Enter an integer: '))
  div(a, b)
except ZeroDivisionError:
  print('Denominator is 0')
print('Performed operation successfully')
`````

**If our program does not catch an exception then Python runtime
catches it.**


`````python
try:
  a = int(input('Enter an integer: '))
  b = int(input('Enter an integer: '))
  div(a, b)
except ValueError:
  print('Invalid literal for int()')
print('Performed operation successfully')
`````

#### `else` block
* The try .. except statement may also have an optional else block.
* If it is present, it must occur after all the except blocks.
* Control goes to else block if no exception occurs during execution of
the try block.

`````python
try:
  a = int(input('Enter an integer: '))
  b = int(input('Enter an integer: '))
  print(div(a, b))
except (ValueError, ZeroDivisionError) as msg:
  print('The error is: ', msg)
except:
  print('A new Error')
else:
  print('The result of division operation between a {a} and b {b} is {a/b}')
  del a, b
`````

#### `finally` Block
* finally block is optional.
* Code in finally always runs, no matter what! Even if a return or
break occurs first.
* finally block is placed after except blocks (if they exist).
* try block must have except block and/or finally block.
* finally block is commonly used for releasing external resources like
files, network connections or database connections, irrespective of
whether the use of the resource was successful or not.

`````python
try:
  a = int(input('Enter an integer: '))
  b = int(input('Enter an integer: '))
  print(div(a, b))
except ZeroDivisionError:
  print('Denominator is 0')
finally:
  print('DOne with division')
`````

`````python
try :
  lst = [10, 20, 'abc', 40, 50]
  for num in lst :
    i = int(num)
    j = i * i
    print(i, j)
except (NameError, ValueError) as err:
  print(err.args)
else:
  print('Total numbers processed', len(lst))
finally:
  print('Done with printing a list')
  del(lst)
`````

* Two things that can be done when an exception is raised:
1. Catch the raised exception object in except block.
2. Raise the exception further.
* If we catch the exception object, we can either perform a graceful exit or rectify the exceptional situation and continue execution.
* If we raise the exception object further - Default exception handler
catches the object, prints stack trace and terminates.

* When exception is raised and except block is executed, control goes
to the next line after except block, unless there is a return or raise in
except block.

An exception workflow can be initiated by using the `raise` statement. To *raise* an exception we need to `raise` an **instance** of an exception type (one that is a subclass of `BaseException`).

**Python raised an `IndexError` exception.**

We can "handle" an exception workflow by using the a `try` statement and handling the exception (if any) in the `except` clause of the handler

`````python
try:
    l[4]
except IndexError as ex:
    print(ex.__class__, ':', str(ex))
`````

Since `IndexError` inherits from `LookupError` which itself inherits from `Exception`, we could actually handle any of those exception types with the same effect

`````python
try:
    l[4]
except LookupError as ex:
    print(ex.__class__, ':', str(ex))
`````

When we handle an exception, the handler will "catch" the exception type we specify, **and any subclass of it**. Broaden our handler to include any subclass of `Exception`

`````python
try:
    l[4]
except Exception as ex:
    print(ex.__class__, ':', str(ex))
`````

`````python
class InsufficientBalanceError(Exception):
  def __init__(self, accno, balance):
    self.__accno = accno
    self.__balance = balance

  def get_info(self):
    return {'Account no': self.__accno,
            'Balance': self.__balance}

class Customer:
  def __init__(self):
    self.__dct = {}
  def add_info(self, accno, name, balance):
    self.__dct[accno] = {'Name': name,
                         'Balance': balance}
  def deposit(self, accno, amount):
    d = self.__dct[accno]
    d['Balance'] += amount
    self.__dct[accno] = d
  def withdraw(self, accno, amount):
    d = self.__dct[accno]
    if d['Balance'] - amount > 1000:
      d['Balance'] -= amount
      self.__dct[accno] = d
    else:
      raise InsufficientBalanceError(accno, d['Balance'])
  def display(self):
    for key, value in self.__dct.items():
      print('Account No: ', key, 'Info: ', value)
    print()
`````

**Write a program that receives 5 integers and stores them and
their cubes in a dictionary. If the number entered is less than 3,
raise a user-defined exception NumberTooSmall, and if the number
entered is more than 30, then raise a user-defined exception
NumberTooBig. Whether an exception occurs or not, at the end**

`````python
class NumberTooSmallError(Exception):
  def __init__(self, num):
    self.num = num
  def get_info(self):
    return {'Number is too small': self.num}

class NumberTooBigError(Exception):
  def __init__(self, num):
    self.num = num
  def get_info(self):
    return {'Number is too big': self.num}

class Numbers:
  def __init__(self):
    self.dct = {}
  def add(self, num):
    self.dct[num] = num*num*num
  def display(self):
    for key, value in self.dct.items():
      print('Number: ', key, 'its cube: ', value)
    print()
`````

`````python
n = Numbers()
print('Enter ten numbers between 3 to 30')
for i in range(5):
  num = int(input('Enter a number: '))
  if num < 3:
    raise NumberTooSmallError(num)
  elif num > 30:
    raise NumberTooBigError(num)
  else:
    n. add(num)
n.display( )
`````
