## Dictionaries

* A dictionary is like a list, but more general. In a list, the index positions have to be integers; in a dictionary, the indices can be (almost) any type.
* You can think of a dictionary as a mapping between a set of indices (which are
called keys) and a set of values. Each key maps to a value. The association of a
key and a value is called a key-value pair or sometimes an item.
* Dictionaries are also known as maps or associative arrays.
* Dictionaries are unordered, indexable and mutable
* A dictionary contains comma separated key : value pairs enclosed
within { }.
* Dictionaries don't allow duplicate keys



### Create a dictionary

To create a dictionary, you place curly brackets ({}) around comma-separated key : value pairs. The simplest dictionary is an empty one, containing no keys or values at all

```python
empty_dict = {}
empty_dict, type(empty_dict)
```

A dictionary contains comma separated key : value pairs enclosed
within { }. Here, 'A101', 'A102', 'B103' are keys, whereas, 'Amol', 'Anil', 'Ravi' are values.

```python
dct = {'A101' : 'Amol', 'A102' : 'Anil', 'B103' : 'Ravi'}
dct
```

Different keys may have same values.

```python
d = {10 : 'A', 20 : 'A', 30 : 'Z'}
d
```

Keys must be unique. If keys are same, but values are different,
latest key value pair getsstored.


```python
d = {10 : 'A', 20 : 'B', 10 : 'Z'} # will store {10 : 'Z', 20 : 'B'}
d
```

If key value pairs are repeated, then only one pair gets stored.

```python
d = {10 : 'A', 20 : 'B', 10 : 'A'} # will store {10 : 'A', 20 : 'B'}
d
```

#### Why dictionaries are called dictionaries?

A dictionary is a way of mapping from one set of arbitrary objects to an associated but equally arbitrary set of objects.

```python
english_to_french = {}
english_to_french['red'] = 'rouge'
english_to_french['blue'] = 'bleu'
english_to_french['green'] = 'vert'
print("red is", english_to_french['red'])
```

### Dictionary Methods

#### Create a new dictionary with fromkeys() method
* The fromkeys() method returns a dictionary with the specified keys and the specified value

```python
# keys for the dictionary
alphabets = {'a', 'b', 'c'}

# value for the dictionary
number = 1

# creates a dictionary with keys and values
dictionary = dict.fromkeys(alphabets, number)

dictionary
```

```python
# keys for the dictionary
alphabets = {'a', 'b', 'c'}

# value for the dictionary
number = [1, 2, 3]

# creates a dictionary with keys and values
dictionary = dict.fromkeys(alphabets, number)

dictionary
```

#### Get an Item by [key] or with get()

**This is the most common use of a dictionary.**

```python
some_pythons = {'Graham': 'Chapman',
                'John': 'Cleese',
                'Eric': 'Idle',
                'Terry': 'Gilliam',
                'Michael': 'Palin',
                'Terry': 'Jones',
                }
some_pythons
```

```python
some_pythons['John']
```

If the key is not present in the dictionary, you’ll get an exception

```python
some_pythons['Groucho']
```

There are two good ways to avoid this. The first is to test for the key at the outset by using in

```python
'Groucho' in some_pythons
```

The second is to use the special dictionary get() function. You provide the dictionary, key, and an optional value. If the key exists, you get its value

```python
some_pythons.get('John')
```

If not, you get the optional value, if you specified one

```python
some_pythons.get('Groucho', 'Not a Key')
```

```python
some_pythons.get('Groucho')
```

#### Remove last inserted key-value pair using popitem() method
The popitem() method removes and returns the (key, value) pair from the dictionary in the Last In, First Out (LIFO) order.

* Returns the latest inserted element (key,value) pair from the dictionary.
* Removes the returned element pair from the dictionary.

```python
some_pythons = {'Graham': 'Chapman',
                'John': 'Cleese',
                'Eric': 'Idle',
                'Terry': 'Gilliam',
                'Michael': 'Palin',
                'Terry': 'Jones',
                }
some_pythons
```

```python
some_pythons.popitem()
```

#### Copy with copy()
- To actually copy keys and values from a dictionary to another dictionary and avoid this, you can use copy()
- This is a shallow copy, and works if the dictionary values are immutable (as they are in this case). If they aren’t, you need deepcopy().

```python
signals = {'green': 'go',
           'yellow': 'go faster',
           'red': 'smile for the camera'}
original_signals = signals.copy()
original_signals
```

```python
signals['blue'] = 'confuse everyone'
```

```python
signals
```

```python
original_signals
```

#### Copy Everything with deepcopy()
Suppose that the value for red in the previous example was a list instead of a single string

```python
signals = {'green': 'go',
           'yellow': 'go faster',
           'red': ['stop', 'smile']}
signals_copy = signals.copy()
signals_copy
```

Let’s change one of the values in the red list

```python
signals['red'][1] = 'sweat'
signals
```

```python
signals_copy
```

The solution is deepcopy()

```python
from copy import copy, deepcopy
```

```python
signals = {'green': 'go',
           'yellow': 'go faster',
           'red': ['stop', 'smile']}
signals_copy = deepcopy(signals)
signals_copy
```

```python
signals['red'][1] = 'sweat'
signals
```

```python
signals_copy
```

#### Return a value using setdefault() method

The setdefault method in python will return:

* The value of the key if it is present in the list
* The function will return None as output if the key is not present in the list and no default value has been set.
* The default value will be returned if it is set by the user and there if the specified key is not present in the list.

```python
d = {1:'one', 2:'two', 3:'three'}
print(d.setdefault(3))
print(d.setdefault(4, 'four'))
print(d.setdefault(5))
print(d)
```

Much like lists and tuples in the previous chapter, dictionaries can be compared with the simple comparison operators == and !=

```python
a = {1:1, 2:2, 3:3}
b = {3:3, 1:1, 2:2}
a == b, a != b
```
