# Python Programming
* Python was invented in the early 1990s to be an accessible, general-purpose programming language.
* As part of that philosophy, it was designed to minimize the amount of **developer time** (the time a programmer actually spends writing code) required to complete a project, rather than to minimize **running time** (the amount of time a program runs).
* This was done, in part, because at the time programmers who needed code that ran fast already had access to the **C and Fortran programming languages, languages which remain the gold standards for speed** even today, but which are relatively hard to work with.
* The design choice was clearly a good one, as it explains much of the success of Python today — yes, **Python may run more slowly than C for most things**, but many programmers are happy to accept that in exchange for being able to write and iterate their Python code much faster than they could in other languages.

##  Why Python?

* Python is a good general-purpose, high-level language.
* The fastest-growing major programming language
* The most popular language for introductory courses
* It is extremely popular in the data science and machine learning worlds.
* Its design makes it very readable, which is more important than it sounds. Being readable also makes it easier to learn and remember; hence, more writable.
* Compared with other popular languages, Python has a gentle learning curve that makes you productive sooner.
*  Python runs almost everywhere and has “batteries included”—a metric boatload of useful software in its standard library.

## Why Not Python?
* Python isn’t the best language for every situation.
*  It’s fast enough for most applications, but it might not be fast enough for some of the more demanding ones. A program written in C, C++, C#, Java, Rust, or Go will generally run faster than its Python equivalent.

Floating-point numbers do not always have precise internal representations. For example, the value 0.1 cannot be exactly represented in binary.

At first glance, 0.1 might seem perfectly accurate, but this is because Python displays numbers in a way that is more human-readable. The actual stored value may differ from what is shown in the output. Since terminal outputs are just strings, Python rounds numbers for better readability, which can obscure their true representation.

````python
0.1
````

To reveal the actual stored value, we can use the format function to specify the number of decimal places displayed. For instance, rounding 0.1 to 25 decimal places makes its inaccuracy visible.

````python
format(0.1, '.25f')
````

Some floats, like 0.125 (which equals 1/8), have exact binary representations. Since 0.125 can be expressed as a finite fraction (1/2³), it is stored precisely.

````python
0.125
````

````python
format(0.125, '.25f')
````

Understanding this concept is crucial when comparing floating-point values. Unlike integers—which maintain exact equality in calculations—floating-point arithmetic can introduce tiny discrepancies.

````python
3 == 1 + 1 + 1
````

For example, adding 0.125 three times results in 0.375, as expected, because both numbers have exact representations. However, performing the same operation with 0.1 leads to a result that may not match expectations. Examining its full string representation reveals why these slight inaccuracies occur.

````python
0.125 + 0.125 == 0.25
````

````python
0.1 + 0.1 + 0.1 == 0.3
````

````python
format(0.1 + 0.1 + 0.1, '.25f')
````

````python
format(0.3, '.25f')
````

Since floating-point numbers don't always have exact representations, direct comparisons using == can lead to unexpected results.

````python
(0.1 + 0.1 + 0.1) - 0.3
````

````python
abs(0.1 + 0.1 + 0.1), abs(0.3)
````

````python
abs(0.1) + abs(0.1) + abs(0.1), abs(0.3)
````

Instead, it's best to compare them using a tolerance level, checking if their difference falls within a small threshold (like 0.01) to account for minor precision errors.

````python
abs((0.1 + 0.1 + 0.1) - 0.3) < 0.001
````

**Why???**

Computers represent numbers using binary, a system based on powers of 2. In binary, the number `1011` converts to decimal as:  
$1 \times 2^0 + 1 \times 2^1 + 0 \times 2^2 + 1 \times 2^3 = 1 + 2 + 0 + 8 = 11$.  

Similarly, floating-point numbers are represented using powers and fractions of 2. For instance,  
$0.5 + 0 + 0.125 = 0.625$  

Some numbers lack finite decimal representations, and the same issue occurs in binary. For example, 0.1 in binary is an infinite repeating fraction, making its exact representation impossible within a finite number of bits.

In Python, **variable assignment follows a clear sequence:**  

1. **Evaluate the right-hand side (RHS)**: Python first computes the expression on the RHS. If it's a simple literal, it directly uses that value. If it's a more complex calculation, Python fully evaluates it before proceeding.

````python
balance = 1000.0 - 50.0  # RHS evaluates to 950.0
circ = 2 * 3.14 * 1.5    # RHS evaluates to 9.42
balance, circ
````

2. **Assign the result to the left-hand side (LHS)**: The evaluated result from the RHS is **linked to the variable name** on the LHS. The name serves as a reference to the computed value rather than storing the value directly.  

**The value stored in a variable can change. Use the assignment operator to give a variable a new value.**

## Operators in Python

1. **Unary Operator**  
   - Works with **one operand**  
   - Usually written before the operand (prefix)  
   - Example: `-x` (negation), `not x` (logical NOT)

2. **Binary Operator**  
   - Operates on **two operands**  
   - Typically appears **between** operands

### Operator Precedence

* Python uses the PEMDAS method for determining order of operations.

**https://docs.python.org/3/reference/expressions.html#operator-precedence**

````python
a = 2
b = 3
c = 4
result = 3 * a ** 3 / (b + 5) + c
print(result)
````

### Type-casting
- Type casting is the process of changing a compatible data-type to the other.
- Like, every datatype can be converted to a string but the reverse is not true.
- Type casting is achieved by passing the datatype to be converted as an argument to the constructer of the data type we want it to be converted to

````python
int('23')
float('23.45')
str(23)
bool(0)
````

````python
a = 10
b = 3.14
c = '78'
d = 'Hello'
````

````python
# convert int to float
print(type(a))
a = float(a)
print(type(a))
````

````python
# convert a str to an int
print(c)
print(type(c))
c = int(c)
print(c)
print(type(c))
````

````python
# convert a str to float
print(d)
print(type(d))
d = float(d)
print(d)
print(type(d))
````
