## Lists
- `Lists`are collections in python which are defined using `[]` and collect various elements inside it(the elements are separated from each other by `,`)
- Like a string, a list is a sequence of values. In a string, the values are characters; in a list, they can be any type ( heterogeneous	sequence	type). The values in list are called elements or sometimes items.
- List is dynamic because based on our requirement we can increase the size and decrease the size.
- Lists are ordered, indexable and mutable
- Lists allow duplicate values

#### Removing values from the list


##### Get an Item by index and Delete It with pop( )
- `pop()` is used to remove a value in the list, based on its index position. If no index is mentioned, pop() removes the last value
- You can get an item from a list and delete it from the list at the same time by using pop().

```python
marxes = ['Groucho', 'Chico', 'Harpo', 'Zeppo']
marxes.pop()
```

```python
marxes.pop(1)
```

```python
marxes
```

```python
li = [1, 2, 3, 4, 55, 6, 7, 8, 9]
# to remove the last value from the list
print('List before removing the value',li)
pop = li.pop()
print('Popped value',pop)
print('List after removing the value',li)
```

```python
# to remove 55 from the list
print('List before removing the value',li)
pop = li.pop(4)
print('Popped value',pop)
print('List after removing the value',li)
```

##### Delete an Item by Value with remove()
- remove() is used to remove a value from the list if the value is known but not its index
- If you’re not sure or don’t care where the item is in the list, use remove() to delete it by value.

```python
marxes = ['Groucho', 'Chico', 'Harpo']
marxes.remove('Groucho')
marxes
```

```python
# removing 6 from the list
li = [1, 2, 3, 4, 55, 6, 7, 8, 9]
print('List before removing the value',li)
li.remove(6)
print('List after removing the value',li)
```

**Note: The pop() method is used to remove values from a list when their index positions are known and remove() is used when their face values are known**

### Reorder Items with sort() or sorted()
* To sort the items in a list by their values rather than their index. Python provides two functions:
> - The list method sort() sorts the list itself, in place.
> - The general function sorted() returns a sorted copy of the list.
+ If the items in the list are numeric, they’re sorted by default in ascending numeric order. If they’re strings, they’re sorted in alphabetical order

```python
marxes = ['Groucho', 'Chico', 'Harpo']
print(f'Original list {marxes} and its identity {id(marxes)}')
sorted_marxes = sorted(marxes)
print(f'Sorted list {sorted_marxes} and its identity {id(sorted_marxes)}')
```

```python
marxes = ['Groucho', 'Chico', 'Harpo']
print(f'Original list {marxes} and its identity {id(marxes)}')
sort_marxes = marxes.sort()
print(f'Sort list {sort_marxes} and its identity {id(sort_marxes)}')
print(f'Original list after sort {marxes} and its identity {id(marxes)}')
```

```python
# to sort a list in an ascending order
li = [1, 3, 2, 4, 5, 3, 1, 2, 4, 1, 3]
print(f'Original list {li} and its addresss is {id(li)}')
li.sort()
print(f'Sorted list {li.sort()} and its addresss is {id(li.sort())}')
print(f'Original list after sort {li} and its identity {id(li)}')
```

```python
# to sort a list in an ascending order
li = [1, 3, 2, 4, 5, 3, 1, 2, 4, 1, 3]
print(f'Original list {li} and its addresss is {id(li)}')
sorted_li = sorted(li)
print(f'Sorted list {sorted_li} and its addresss is {id(sorted_li)}')
```

**The default sort order is ascending, but you can add the argument reverse=True to set it to descending:**

```python
# to sort a list in descending order
li = [1, 3, 2, 4, 5, 3, 1, 2, 4, 1, 3]
print(f'Original list {li} and its addresss is {id(li)}')
print(f'Sorted list {li.sort(reverse=True)} and its addresss is {id(li.sort(reverse=True))}')
print(f'Original list after sort {li} and its identity {id(li)}')
```

```python
# to sort a list in descending order
li = [1, 3, 2, 4, 5, 3, 1, 2, 4, 1, 3]
print(f'Original list {li} and its addresss is {id(li)}')
sorted_li = sorted(li, reverse = True)
print(f'Sorted list {sorted_li} and its addresss is {id(sorted_li)}')
```
