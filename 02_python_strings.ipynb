# Strings

## What is a String?

- A string is a sequence of characters or Unicode characters.
- Computers do not deal with characters, they deal with numbers (binary). Even though you may see characters on your screen, internally it is stored and manipulated as a combination of 0's and 1's.
- This conversion of character to a number is called encoding, and the reverse process is decoding. ASCII and Unicode are some of the popular encoding used.
- Like many other popular programming languages, strings in Python are arrays of bytes representing unicode characters.

### String Slicing / String Indexing

- The index must be an integer. We can't use float or other types, this will result into TypeError.

- step: how many elements to skip, when accessing a group of elements

#### Formatting numbers
- d: specify how many digits are required
- f: floating point number(float)
- b: Binary format
- o: Octal Format
- x: Hexa Decimal Format

```python
dec = 123
print(f'The decimal number is: {dec:d}')
print(f'The decimal number is: {dec:2d}')
print(f'The decimal number is: {dec:4d}')

# Fill the blank spaces with zeros
print(f'The decimal number is: {dec:04d}')
```

```python
flt = 3/18 # The default precision is 6
print(f'The floating point number is: {flt}')
print(f'The floating point number is: {flt:f}')
print(f'The floating point number is: {flt:.2f}')

# It takes a float argument and assigns a minimum width of 10 including "." and after decimal  point excatly 4 digits are allowed with round operation if required
print(f'The floating point number is: {flt:10.4f}')
# Fil the blank spaces with zeros
print(f'The floating point number is: {flt:012.8f}')
```

```python
dec = 123
print(f"Binary Form:{dec:b}")
print(f"Octal Form:{dec:o}")
print(f"Hexa decimal Form:{dec:x}")
print(f"Hexa decimal Form:{dec:X}")
```

#### Formatting signed numbers

* Negative sign prints automatically but if we want to type '+' sign for positive numbers the {:+d} and {:+f} are used

```python
print(f"Positive integer with sign:{dec:+d}")
print(f"Negative integer with sign:{-dec:d}")
print(f"Positive floating point number with sign:{flt:+f}")
print(f"Negative floating point number with sign:{-flt:f}")
```

#### Number alignment
* <,>,^ and = are used for alignment
* Default Alignment for numbers is Right Alignment.

```python
print('Default alignment: {:d}'.format(123))
print('Default alignment: {:d}'.format(-123))
print('Default alignment: {:5d}'.format(123))
print('Default alignment: {:010d}'.format(123))
```

```python
# Left alignment
print('Left alignment: {:<d}'.format(123))
print('Left alignment: {:<d}'.format(-123))
print('Left alignment: {:<5d}'.format(123))
print('Left alignment: {:<010d}'.format(123))
```

```python
# Right alignment
print('Right alignment: {:>d}'.format(123))
print('Right alignment: {:>d}'.format(-123))
print('Right alignment: {:>5d}'.format(123))
print('Right alignment: {:>010d}'.format(123))
```

```python
# Centre alignment
print('Centre alignment: {:^d}'.format(123))
print('Centre alignment: {:^d}'.format(-123))
print('Centre alignment: {:^5d}'.format(123))
print('Centre alignment: {:^010d}'.format(123))
```

### trims whitespace
- lstrip( ) - removes whitespace from the left of string including \t.
- rstrip( ) - removes whitespace from the right of string including \t.

```python
va1 = '       How are you?\n'
print(va1)
va2 = 'How are you?'
print(va2)
```

#### lstrip( )
- removes whitespace from the left of string including

```python
var3 = '                Hello                      '
print(var3)
var3 = var3.lstrip()
print(var3)
```

#### rstrip( )
- removes whitespace from the right of string including \t.

```python
var3 = '                Hello                      '
print(var3)
var3 = var3.rstrip()
print(var3)
```

```python
var3 = '             Hello  $$$$               '
print(var3)
var3 = var3.rstrip(' $')
print(var3)
```
